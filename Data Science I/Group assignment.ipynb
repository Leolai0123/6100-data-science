```python
import pandas as pd
```

```python

data = pd.read_csv('https://raw.githubusercontent.com/Leolai0123/6100-data-science/main/Data.csv')
```

```python
# 计算每个location的Alcohol列的平均值
location_alcohol_avg = data.groupby('LocationAbbr')['DataValue'].mean().reset_index()

location_alcohol_avg
```

```python
chronic = pd.read_csv('https://raw.githubusercontent.com/Leolai0123/6100-data-science/main/Chronic%20kidney%20disease.csv')
```

```python
location_ckd_avg = chronic.groupby('LocationAbbr')['DataValue'].mean().reset_index()

location_ckd_avg
```

```python
import pandas as pd
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#relationship between response rate and clv_monthly
plt.figure(figsize=(10, 6))
sns.regplot(x=location_alcohol_avg['DataValue'], y=location_ckd_avg['DataValue'])
plt.xlabel('Alcohol')
plt.ylabel('CKD')
plt.title('Relationship ')
plt.tight_layout()
plt.show()
```
